# 13 Exploratory Data Analysis

### Titanic Survivor Data

In this dataset, the details of a number of Titanic passengers were documented. In addition to the passengers' names, their ticket class, age, sex, number of siblings/spuses aboard, number of parents/children aboard, ticket number, fare, cabin number, port of embarkation, and most importantly whether or not they survived were indicated. This dataset comes from historical documentation, and it is particularly interesting because it allows us to determine what factors had a role in passenger survival. In this basic analysis, a few factors will be looked at that indicated higher rates of survival. 



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./train.csv') # read data

Survived = data.loc[data['Survived']==1] # create survived data set
NSurvived = data.loc[data['Survived']==0] # create died data set

gens = Survived['Sex'].value_counts() # count female and male survivors
pclass = Survived['Pclass'].value_counts() # count class survivors

fare = pd.concat([NSurvived['Fare'], Survived['Fare']], axis=1, ignore_index=True) # create Fare vs survival dataset

# Plot Age vs Sib Spouse in two survivor/non survivor groups with size parent children aboard
ax = Survived.plot.scatter(x='Age', y='SibSp', s=data['Parch']*20, color='DarkBlue', alpha=0.5, label='Survived');
NSurvived.plot.scatter(x='Age', y='SibSp', s=data['Parch']*20, color='Red', alpha=0.5, label='Died', ax=ax, 
                       figsize=(12, 4));
plt.xlabel('Age'), plt.ylabel('Siblings and Spouses Aboard')
plt.title('Age vs. Siblings and Spouses, Parents and Children Represented by Size')
plt.show();

# Fare vs Survival boxplot
fare.plot.box(sym='');
plt.xlabel('Survival'), plt.ylabel('Fare'), plt.title('Fare Boxplot')
plt.show();

# Pie chart survivor sex
gens.plot.pie(figsize=(6, 6)); 
plt.title('Survivor Sex')
plt.show();

# Pie chart survivor ticket class
pclass.plot.pie(figsize=(6, 6));
plt.title('Survivor Ticket Class')
plt.show();
```

Looking at the data, a few conclusions can be drawn. From the first plot, we can see that generally children had a higher survival rate than adults. However, this survival was mitigated by an increase in siblings and spouses on board, which extends across all ages. This phenomenon could be associated with a drive to save or find family, dying in the process.

The boxplot shows fares of survivors and non-survivors. The range of fares of survivors is much larger than non-survivors, but it is clear that the average fare of survivors was higher. This indicates that paying a higher fare ultimately led to a better chance of survival.

The two pie charts show demographics of the group of survivors. The fraction of female survivors is much larger than male survivors, further supporting the idea that women and children were likely to be saved first. The ticket class is less clear, with fractions of each class similarly represented, and shows that ticket class may not have been a good indicator of survival
